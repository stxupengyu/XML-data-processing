```python
#上面已经实现了转换单个xml文件，下面把该过程写成函数，从而能把它应用到每个xml文件
def extract(name):
    tree = ET.ElementTree(file=name)
    root = tree.getroot()#找到根节点
    #root.tag, root.attrib
    record = []   
    for child_of_root in root:#遍历第一层
        if(child_of_root.tag=='讨论'):
            for child1 in child_of_root:#遍历‘讨论’这一层
                record2=[]
                for child2 in child1:#遍历每个‘item’层
                    record2.append(child2.text)#把每个‘item’层的文本保存在record2
                record.append(record2)#吧每个record2保存到record
    result=pd.DataFrame(record)
    return result
```

```python
#获取当前文件夹的文件名，根据文件名调用函数并且合并dataframe对象
import os
file_dir=os.getcwd()#获取当前工作目录路径
namelist=os.listdir(file_dir)#获取当前工作目录所有文件名
data=extract('国潮下热门5月_42699654_1592809651097.xml')
for name in namelist[-40:]:#-10即为遍历10个xml文件
    data=pd.concat([data,extract(name)])
data=data.drop_duplicates()
data=pd.DataFrame(data.values)
data.tail()
```
